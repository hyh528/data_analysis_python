```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

## 기본 그래프(plot)

```python
plt.plot([1,2,3,4,5,6,5,4,3,2,1])
plt.show()
```

```python
## sin 함수 그려보기
x = np.arange(0,12,0.01)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

## Title 설정

```python
plt.plot(x,y)
plt.title('sin graph')
plt.show()
```

## Grid 설정

```python
plt.plot(x,y)
plt.grid()
plt.show()
```

## X,Y 라벨

```python
plt.plot(x,y)
plt.plot(x,y-1)
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.show()
```

## 범례(legend)

```python
plt.plot(x, np.sin(x), label='sin')
plt.plot(x, np.cos(x), label='cos')
plt.legend(loc=(1, 0.1))
plt.show()
```
